```python
import pandas as pd
from datetime import datetime
```

```python
d = {
'EmpID'  : [1000,1001,1002,1003],
'EmpName': ['Samsheer','Inaya','Ruby','Aydin'],
'Age'    : [39,5,37,9]
}
idx = pd.date_range(start='2024-01-01', periods=4, freq='3ME')
columns=['EmpID','EmpName','Age']
df = pd.DataFrame(d, index = idx, columns=columns)
df
```

```python
d = [
[1000,'Samsheer',39],
[1001,'Inaya',5],
[1002,'Ruby',37],
[1003,'Aydin',9]
]
idx = pd.date_range(start='2024-01-01', periods=4, freq='3ME')
columns=['EmpID','EmpName','Age']
df = pd.DataFrame(d, index = idx, columns=columns)
df
```

```python
tuples = [(1, 'Samsheer'),(1, 'Inaya'),(2, 'Ruby'),(2,'Aydin')]
idx = pd.MultiIndex.from_tuples(tuples, names=('number', 'Name'))
columns=['EmpID','EmpName','Age']
df = pd.DataFrame(d, index = idx, columns=columns)
df
```

```python
#Multi Index
tuples = [
    ('cobra', 'mark i'),
    ('cobra', 'mark ii'),
    ('sidewinder', 'mark i'),
    ('sidewinder', 'mark ii'),
    ('viper', 'mark ii'),
    ('viper', 'mark iii')
]
index = pd.MultiIndex.from_tuples(tuples)
values = [
    [12, 2],
    [0, 4],
    [10, 20],
    [1, 4],
    [7, 1],
    [16, 36]]

df = pd.DataFrame(values, columns=['max_speed', 'shield'], index=index)

df

```
