```python
import pandas as pd
import cx_Oracle
import sqlalchemy
from sqlalchemy.exc import SQLAlchemyError

try:
    engine = sqlalchemy.create_engine("oracle+cx_oracle://system:oracle@localhost:1522/?service_name=orclpdb", arraysize=1000)
    qry_sql = "select * from HR.EMPLOYEES where employee_id =100"
    df_sql = pd.read_sql(qry_sql, engine)
except SQLAlchemyError as e:
    print(e)
```

```python
def dbquery(query_db, cred_db):
    curs = cred_db.cursor()
    curs.execute(query_db)
    rows = curs.fetchall()
    curs.close()
    return rows

connectString= cx_Oracle.makedsn('localhost', 1522, 'ORCL19C')
DB_CRED = cx_Oracle.connect('system', 'oracle', connectString)
QUERY = "select * from all_tables where upper(owner)=upper('SYS') and upper(table_name) like upper('%MAP_ELEMENT$%')"
RES = dbquery(QUERY, DB_CRED)
```
