```python
import pandas as pd
```

```python
df = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3', 'K4', 'K5'],
                   'A': ['A0', 'A1', 'A2', 'A3', 'A4', 'A5']})

other = pd.DataFrame({'key': ['K0', 'K1', 'K2'],
                      'B': ['B0', 'B1', 'B2']})
```

```python
df
```

```python
df.set_index('key')
```

```python
other
```

```python
other.set_index('key')
```

```python
df.join(other, lsuffix='_caller', rsuffix='_other')
```

```python
df.set_index('key').join(other.set_index('key'), on='key')
```

```python
df.join(other.set_index('key'), on='key')
```

```python
df = pd.DataFrame(
    {'key': ['K0', 'K1', 'K1', 'K3', 'K0', 'K1'],
    'A': ['A0', 'A1', 'A2', 'A3', 'A4', 'A5']
    })
```
